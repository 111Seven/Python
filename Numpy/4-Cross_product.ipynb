# Cross Product
Similar to dot product, cross product can be defined between two vectors, vector and tensor, tensor and tensor, as long as the dimensions are matching. Now let's define couple of vectors in 3 dimensional cartesian coordinates. For out convenience let's make their origin as (0,0,0). The vectors are $$ a = \begin{bmatrix} 5 \\ 1 \\ 4 \end{bmatrix} $$ $$ b = \begin{bmatrix} 5 \\ 7 \\ 8 \end{bmatrix} $$

```python
#Import required modules
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#define the vectore
a = np.array([[5],[1],[4]])
b = np.array([[5],[7],[8]])
print("a :{}".format(a))
print("b :{}".format(b))
```

```python
#Now lets plot the vectors
fig = plt.figure(figsize=(8,8))
ax  = fig.add_subplot(projection="3d")
#defining the x,y,z coordinates to plot a
xa = np.array([0,a[0,0]])
ya = np.array([0,a[1,0]])
za = np.array([0,a[2,0]])
#defining the x,y,z coordinates to plot b
xb = np.array([0,b[0,0]])
yb = np.array([0,b[1,0]])
zb = np.array([0,b[2,0]])
#lets plot the vectors
ax.plot(xa,ya,za,color="b") #plotting the vector a in blue
ax.scatter(a[0,0],a[1,0],a[2,0],color="b",marker=">") #putting the arrow for vector a
ax.plot(xb,yb,zb,color="g") #plotting the vector b in green
ax.scatter(b[0,0],b[1,0],b[2,0],color="g",marker=">") #putting the arrow for vector a
```

## Definition of cross product
The definition of cross product between two vectors is given by $$ a \, \times \, b = \begin{bmatrix} (a_2b_3 - a_3b_2) & -(a_1 b_3 -b_1 a_3) & (a_1b_2 - a_2b_1) \end{bmatrix} $$ In our example we can write cross product as $$ a \, \times \, b = \begin{bmatrix} (1 \times 8 - 4 \times 7) & -(5 \times 8 - 5 \times 4) & (5 \times 7 - 1 \times 5) \end{bmatrix} $$  $$ a \, \times \, b =  \begin{bmatrix} -20 & -20 & 30 \end{bmatrix} $$ The vector $ a \times b $ is in the direction $ \perp $ to the plane containing the vectors $ a $ and $ b $

```python
#Now lets find the same using numpy 
ab_cross = np.cross(a,b,axisa=0,axisb=0,axisc=0) #row vectors are expected as input
ab_cross
```

```python
#Now lets plot this vector in the 
fig = plt.figure(figsize=(8,8))
ax  = fig.add_subplot(projection="3d")
#defining the x,y,z coordinates to plot a
xa = np.array([0,a[0,0]])
ya = np.array([0,a[1,0]])
za = np.array([0,a[2,0]])
#defining the x,y,z coordinates to plot b
xb = np.array([0,b[0,0]])
yb = np.array([0,b[1,0]])
zb = np.array([0,b[2,0]])
#defining the x,y,z coordinates to plot a x b
xab = np.array([0,ab_cross[0,0]])
yab = np.array([0,ab_cross[1,0]])
zab = np.array([0,ab_cross[2,0]])
#lets plot the vectors
ax.plot(xa,ya,za,color="b") #plotting the vector a in blue
ax.scatter(a[0,0],a[1,0],a[2,0],color="b",marker=">") #putting the arrow for vector a
ax.plot(xb,yb,zb,color="g") #plotting the vector b in green
ax.scatter(b[0,0],b[1,0],b[2,0],color="g",marker=">") #putting the arrow for vector a
ax.plot(xab,yab,zab,color="r") #plotting the vector a x b in red
ax.scatter(ab_cross[0,0],ab_cross[1,0],ab_cross[2,0]-0.1,
           color="r",marker="<") #putting the arrow for vector a x b
```

## Angle between two vectors
The magnitude of the cross product can also be find by $$ |a \times b| = |a| |b| sin \theta $$ Now from here $$ \theta = sin^{-1} \left(\frac{| a \times b |}{|a| |b|} \right)$$
where $ |.| $ denotes the magnitude of the vector

```python
#Now the angle between vectors
ab_cross_mag = np.linalg.norm(ab_cross)
a_mag = np.linalg.norm(a)
b_mag = np.linalg.norm(b)
theta = np.arcsin(ab_cross_mag/(a_mag*b_mag))
theta*180/np.pi
```

## Area of parallelogram
Let's consider that vector $a$ and $b$ has common origin point as in our case (0,0,0). Then we can take $a$ and $b$ are the sides of the parallelogram. The area of the parallelogram is given by $$ A_P = | a \times b| $$ The vector $$ n_A = \frac{a \times b}{|a| |b|} $$ denotes the unit vector normal to the plane of the parallelogram

```python
#Now let's find the area of the parallelogram
ab_cross_mag
```

```python
#Now let's find the unit normal vector to the parallelogram
n_A = ab_cross/ab_cross_mag
n_A
```

```python
#Now lets plot this vector in the 
fig = plt.figure(figsize=(6,6))
ax  = fig.add_subplot(projection="3d")
#defining the x,y,z coordinates to plot a
xa = np.array([0,a[0,0]])
ya = np.array([0,a[1,0]])
za = np.array([0,a[2,0]])
#defining the x,y,z coordinates to plot b
xb = np.array([0,b[0,0]])
yb = np.array([0,b[1,0]])
zb = np.array([0,b[2,0]])
#defining the x,y,z coordinates to plot a x b
xnab = np.array([0,n_A[0,0]])
ynab = np.array([0,n_A[1,0]])
znab = np.array([0,n_A[2,0]])
#lets plot the vectors
ax.plot(xa,ya,za,color="b") #plotting the vector a in blue
ax.scatter(a[0,0],a[1,0],a[2,0],color="b",marker=">") #putting the arrow for vector a
ax.plot(xb,yb,zb,color="g") #plotting the vector b in green
ax.scatter(b[0,0],b[1,0],b[2,0],color="g",marker=">") #putting the arrow for vector a
ax.plot(xnab,ynab,znab,color="r") #plotting the vector a x b in red
ax.scatter(n_A[0,0],n_A[1,0],n_A[2,0],
           color="r",marker="<") #putting the arrow for vector a x b
```
