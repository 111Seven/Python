# Dot product
Dot product can be defined between two vectors, vector and tensor, tensor and tensor. As long as their dimensions are in agreement we can define a dot product between them. Now let's quickly define couple of vectors in 3 dimensional cartesian coordinate system. For our convenience let's make their origin as (0,0,0). The vectors are $$ a = \begin{bmatrix} 5 \\ 1 \\ 4 \end{bmatrix} $$ $$ b= \begin{bmatrix} 5 \\ 7 \\ 8 \end{bmatrix} $$

```python
# Import the required modules 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Define the vectors 
a = np.array([[5],[1],[4]])
b = np.array([[5],[7],[8]])
print("a: {}".format(a))
print("b: {}".format(b))
```

```python
#Now lets plot the vectors in 3d space
#lets define the axes
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
#ax  = fig.gca(projection="3d")  
xs = np.array([0,a[0,0]])
ys = np.array([0,a[1,0]])
zs = np.array([0,a[2,0]])
xs1 = np.array([0,b[0,0]])
ys1 = np.array([0,b[1,0]])
zs1 = np.array([0,b[2,0]])
ax.plot(xs,ys,zs,color="b")   #blue color for vector a
ax.scatter(a[0],a[1],a[2],marker=">",color="b")   #blue color for vector a
ax.plot(xs1,ys1,zs1,color="g")#green color for vector b
ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
```

## Definition of dot product
Dot product between two vectors is defined as $$ a.b = a_1b_1+ a_2b_2 + a_3b_3 $$ From here we can write the dot product $$ a.b = 5*5 + 1*7 + 4*8 $$ $$ a.b = 64$$

```python
#Now the dot product between these two vecotors is given by 
ab_dot = np.dot(a.T,b)
ab_dot
```

## Angle between two vectors
Dot product is also defined as $$ a.b = |a| |b| cos \theta $$ Now from here the angle between two vectors is given by $$ \theta = cos^{-1} \left( \frac{a.b}{|a||b|} \right) $$ where $|a|$ and $|b|$ are the magnitudes of the vectors which are given by $$ |a| = \sqrt{(a_1^2 +a_2^2 + a_3^2)} $$ $$ |b| = \sqrt{(b_1^2 +b_2^2 + b_3^2)} $$

```python
#The magnitude of a vector
a_mag = np.linalg.norm(a)
a_mag
```

```python
#The magnitude of b vector
b_mag = np.linalg.norm(b)
b_mag
```

```python
#the angle between two vectors is given by
theta = np.arccos(ab_dot/(a_mag*b_mag))
theta*180/np.pi #theta in degrees
```

## Projection of a vector
From the above we can define the projection of vector $a$ on $b$. It is defined as $$ P_a = (|a| cos \theta) \frac{b}{|b|} $$ 
Similarly projection of vector $b$ on $a$ is given by $$ P_b = (|b| cos \theta ) \frac{a}{|a|} $$

```python
#Projection of a on b
P_a = (a_mag*np.cos(theta))*(b)/b_mag
P_a
```

```python
#Lets plot the projection of a on b
#lets define the axes
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
#ax  = fig.gca(projection="3d")
#x, y, z's to draw line a 
xs = np.array([0,a[0,0]])
ys = np.array([0,a[1,0]])
zs = np.array([0,a[2,0]])
#x, y, z's to draw line b
xs1 = np.array([0,b[0,0]])
ys1 = np.array([0,b[1,0]])
zs1 = np.array([0,b[2,0]])
#x, y, z's to draw projected line on a 
psx = np.array([0,P_a[0,0]])
psy = np.array([0,P_a[1,0]])
psz = np.array([0,P_a[2,0]])
#x, y, z's to show projection
psx1  = np.array([a[0,0],P_a[0,0]])
psy1  = np.array([a[1,0],P_a[1,0]])
psz1  = np.array([a[2,0],P_a[2,0]])
#plotting the vectors
ax.plot(xs,ys,zs,color="b")   #blue color for vector a
ax.scatter(a[0],a[1],a[2],marker=">",color="b")   #blue color for vector a
ax.plot(xs1,ys1,zs1,color="g")#green color for vector b
ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
ax.plot(psx,psy,psz,color="r")#red color to show the projection of a on b
ax.plot(psx1,psy1,psz1,color="k",ls="dashed")#black color for showing projection
#ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
```

```python
#Projection of b on a
P_b = (b_mag*np.cos(theta))*(a)/a_mag
P_b
```

```python
#Lets plot the projection of b on a
#lets define the axes
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
#ax  = fig.gca(projection="3d")
#x, y, z's to draw line a 
xs = np.array([0,a[0,0]])
ys = np.array([0,a[1,0]])
zs = np.array([0,a[2,0]])
#x, y, z's to draw line b
xs1 = np.array([0,b[0,0]])
ys1 = np.array([0,b[1,0]])
zs1 = np.array([0,b[2,0]])
#x, y, z's to draw projected line on a 
psx = np.array([0,P_b[0,0]])
psy = np.array([0,P_b[1,0]])
psz = np.array([0,P_b[2,0]])
#x, y, z's to show projection
psx1  = np.array([b[0,0],P_b[0,0]])
psy1  = np.array([b[1,0],P_b[1,0]])
psz1  = np.array([b[2,0],P_b[2,0]])
#plotting the vectors
ax.plot(psx,psy,psz,color="r")#red color for vector b projected on a
ax.plot(xs,ys,zs,color="b")   #blue color for vector a
ax.scatter(a[0],a[1],a[2],marker=">",color="b")   #blue color for vector a
ax.plot(xs1,ys1,zs1,color="g")#green color for vector b
ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
ax.plot(psx1,psy1,psz1,color="k",ls="dashed")#black color for showing projection
#ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
```
