# Dot product
Dot product can be defined between two vectors, vector and tensor, tensor and tensor. As long as their dimensions are in agreement we can define a dot product between them. Now let's quickly define couple of vectors in 3 dimensional cartesian coordinate system. For our convenience let's make their origin as (0,0,0). The vectors are $$ a = \begin{bmatrix} 2 \\ 1 \\ 4 \end{bmatrix} $$ $$ b= \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} $$

```python
# Import the required modules 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Define the vectors 
a = np.array([[3],[1],[4]])
b = np.array([[5],[2],[8]])
print("a: {}".format(a))
print("b: {}".format(b))
```

```python
#Now lets plot the vectors in 3d space
#lets define the axes

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
#ax  = fig.gca(projection="3d")  
xs = np.array([0,a[0,0]])
ys = np.array([0,a[1,0]])
zs = np.array([0,a[2,0]])
xs1 = np.array([0,b[0,0]])
ys1 = np.array([0,b[1,0]])
zs1 = np.array([0,b[2,0]])
ax.plot(xs,ys,zs,color="b")   #blue color for vector a
ax.scatter(a[0],a[1],a[2],marker=">",color="b")   #blue color for vector a
ax.plot(xs1,ys1,zs1,color="g")#green color for vector b
ax.scatter(b[0],b[1],b[2],marker=">",color="g")   #green color for vector b
```
