# Dot and Cross product
Let's say we have three vectors $a$, $b$, and $c$. Then we can perform dot and cross product on all the three vectors. The quantity represents the volume of the parallelopiped $V$ formed by these three vectors. $$ V = a . (b \times C) $$
Let's say the vectors are $$ a = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$ $$ b = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix} $$ $$ c  = \begin{bmatrix} c_1 \\ c_2 \\ c_3 \end{bmatrix} $$ 

Now volume of the parallelopiped formed by these three vectors is given by $$ V = a_1(b_2c_3 - b_3c_2) - a_2(b_1c_3-b_3c_1) + a_3(b_1c_2-b_2c_1)$$

```python
#Import all the required modules
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#Now let's see this by any example
a = [1,-1,2]
a = np.array(a).reshape(-1,1)
b = [2, 1,3]
b = np.array(b).reshape(-1,1)
c = [3, 2,-1]
c = np.array(c).reshape(-1,1)
a.shape
```

```python
#function to plot a vector
def vector_plot(a, ax, col):
    xa = np.array([0,a[0,0]])    
    ya = np.array([0,a[1,0]])
    za = np.array([0,a[2,0]])
    ax.plot(xa, ya, za, color = col)
    ax.scatter(a[0,0],a[1,0],a[2,0],color = col, marker = "o")
```

```python
#Now lets plot the vectors
fig = plt.figure(figsize=(8,8))
ax  = fig.add_subplot(projection="3d")
vector_plot(a, ax, "r")
vector_plot(b, ax, "g")
vector_plot(c, ax, "b")
```

```python
#Now lets find the crass product of the vectors
b_c_cross = np.cross(b,c, axisa=0, axisb=0, axisc=0)
b_c_cross
```

```python
#Now lets find the volume of parallelopiped
V = np.dot(a.T, b_c_cross)
V
```
